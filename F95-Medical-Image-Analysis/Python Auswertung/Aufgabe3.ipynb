```python
sampl = [0.05, 0.1, 0.2, 0.4, 0.5, 0.8, 1]
mi =    [-0.522792, -0.519224,-0.522792, -0.6373, -0.6275, -0.63927, -0.639169]
time =  [7.93192, 13.6309, 21.1612, 39.686, 41.3281, 75.3088, 82.8692]
```

```python
len(mi)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(sampl, mi)
```

```python
fig, ax1 = plt.subplots()
fig.suptitle('Variation Sampling-Prozentzahl [4 Level, Random]', fontsize=12)

ax1.set_xlabel('Sampling Percentage bei Strategie Random')
ax1.set_ylabel('Wert der Mutual Information Metrik [a.u.]', color='green')
ax1.scatter(sampl, mi, color='green', marker='+')
ax1.tick_params(axis='y', labelcolor='green')
ax1.set_xlim([-.01, 1.01])

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Zeit der Registrierung [sek]', color='blue')  # we already handled the x-label with ax1
ax2.scatter(sampl, time, color='blue', marker='s')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_xlim([-.01, 1.01])

#Plot abspeichern
plt.savefig('teil3-variation.pdf', dpi=300, format='pdf')

#fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
