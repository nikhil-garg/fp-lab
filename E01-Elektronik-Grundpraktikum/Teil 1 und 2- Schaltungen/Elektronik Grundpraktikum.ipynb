# Elektronik Grundpraktikum

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
# Messwerte einlesen
#Einlesen der Daten aus Tabelle 1
f, U_E, dU_E, U_A, dU_A = np.loadtxt('tab1.txt', skiprows=1, usecols=(0,1,2,3,4), unpack=True)
```

```python
V_U = U_A / U_E
dV_U = V_U*np.sqrt((dU_A/U_A)**2+(dU_E/U_E)**2)
```

```python
dV_U
```

```python
plt.yscale('log')
plt.ylim((5,20))
plt.xlim((-10, 200))
plt.errorbar(f, V_U, marker='x', yerr=dV_U, linestyle='--')
plt.xlabel('Frequenz $f$ in Hz')
plt.ylabel('Verstärkung $V_U$')
plt.title('Plot 1: Grenzfrequenz Emitterschaltung')
#plt.savefig('plot01.pdf',format='pdf')
```

```python
plt.yscale('log')
plt.xscale('log')
plt.ylim((5,20))
plt.xlim((10, 1800000))
plt.errorbar(f, V_U, marker='x', yerr=dV_U, linestyle='--')
plt.xlabel('Frequenz $f$ in Hz')
plt.ylabel('Verstärkung $V_U$')
plt.title('Plot 2: Frequenzgang Emitterschaltung')
#plt.savefig('plot02.pdf',format='pdf')
```

```python
# Werte in Ausgabedatei schreiben:
with open('daten1_achtung_neu_berechnet.txt', 'w') as file:
    file. write('f	U_E	dU_E	U_A 	dU_A	V_U	dV_U\n')
    for i in range(len(f)):
        file.write(str(f[i])[:-2]+'	'+
                   str(U_E[i])+'	'+
                   str(dU_E[i])+'	'+
                   str(U_A[i])+'	'+
                   str(dU_A[i])+'	'+
                   str(round(V_U[i],2))+'	'+
                   str(round(dV_U[i],2))+'\n')
```
